## All-time scrobbles

```python
# read the all-time scrobbles data set
scrobbles = pd.read_csv('data/lastfm_scrobbles.csv', encoding='utf-8')
scrobbles = scrobbles.drop('timestamp', axis=1)
print('{:,} total scrobbles'.format(len(scrobbles)))
print('{:,} total artists'.format(len(scrobbles['artist'].unique())))
```

```python
scrobbles.head()
```

```python
# convert to datetime
scrobbles['timestamp'] = pd.to_datetime(scrobbles['datetime'])
```

```python
scrobbles = scrobbles.drop(labels=['datetime'], axis=1)
```

```python
scrobbles['year'] = scrobbles['timestamp'].dt.year
scrobbles['month'] = scrobbles['timestamp'].dt.month
scrobbles['day'] = scrobbles['timestamp'].dt.day
scrobbles['hour'] = scrobbles['timestamp'].dt.hour
scrobbles['weekday_name'] = scrobbles['timestamp'].dt.strftime("%A") 
```

```python
scrobbles.head()
```

## Songs by day week

```python
# get the play count sum by day of the week
dayname_counts = scrobbles['weekday_name'].value_counts().to_frame('counts').rename_axis('day').reset_index()
```

```python
dayname_counts
```

```python
dayname_counts.values
```

```python
alt.Chart(dayname_counts).mark_bar().encode(
    x=alt.X('day:N',title='count'),
    y=alt.Y('counts:Q')
    
).properties(
    width=600,
    title = 'Songs by day'
    
).configure_title(
    align='right',
    fontSize=20,
    font='Courier',
    anchor='start',
    color='gray',
    
)
```

```python
weekday_hour_counts = scrobbles.groupby(['weekday_name','hour']).count()['track'].to_frame()
#hour_numbers = weekday_hour_counts.index.levels[1]
#day_numbers = weekday_hour_counts.index.levels[0]
#day_names = {0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday', 4:'Friday', 5:'Saturday', 6:'Sunday'}b
```

```python
weekday_hour_counts.head()
```

```python
weekday_hour_counts = weekday_hour_counts.reset_index()
```

```python
weekday_hour_counts.columns
```

```python
alt.Chart(weekday_hour_counts).mark_line().encode(
    x='hour',
    y='track',
    color='weekday_name',
    strokeDash='weekday_name',
).properties(
    width=600,
)
```

```python
alt.Chart(weekday_hour_counts).mark_area().encode(
    alt.X('hour',
        #axis=alt.Axis(format='%h', domain=False, tickSize=0)
    ),
    alt.Y('track', stack='center', axis=None),
    alt.Color('weekday_name:N',
        #scale=alt.Scale(scheme='category20b')
    )
).properties(
    width=600
).interactive()
```
